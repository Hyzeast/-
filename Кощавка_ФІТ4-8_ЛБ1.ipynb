```python
#Створіть одновимірний масив (вектор) з простими числами, не перевищуючи 70, та виведіть його значення.

# Функція для генерації списку простих чисел менше або рівних заданому ліміту
def generate_primes(limit):
    primes = []
    for num in range(2, limit + 1):
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
    return primes

# Генерація простих чисел, що не перевищують 70
prime_numbers = generate_primes(70)

# Виведення значень простих чисел
print(prime_numbers)
```

```python
#Створіть матрицю розміром 2x3 та вектор розміром 3x1, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими дійсними 
# числами від 0 до 1, і знайдіть суму елементів матриці, які менше за відповідний елемент вектору.

import numpy as np

# Створення матриці 2x3 з випадковими дійсними числами від 0 до 1
matrix = np.random.rand(2, 3)
print("Матриця 2x3:")
print(matrix)

# Створення вектора 3x1 з випадковими дійсними числами від 0 до 1
vector = np.random.rand(3, 1)
print("Вектор 3x1:")
print(vector)

# Знаходження суми елементів матриці, які менше за відповідний елемент вектору
sum_elements = np.sum(matrix < vector.T, axis=1)
print("Сума елементів матриці, які менше за відповідний елемент вектору:")
print(sum_elements)
```

```python
# Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -5 до 5 та знайдіть добуток елементів матриці, 
# які більше за їх середнє арифметичне значення.
import numpy as np

# Створення матриці 3x3 з випадковими дійсними числами від -5 до 5
matrix = 10 * np.random.rand(3, 3) - 5
print("Матриця 3x3:")
print(matrix)

# Обчислення середнього арифметичного значення елементів матриці
mean_value = np.mean(matrix)
print("Середнє арифметичне значення елементів матриці:")
print(mean_value)

# Знаходження елементів, які більше за середнє арифметичне значення
greater_than_mean = matrix[matrix > mean_value]
print("Елементи, які більше за середнє арифметичне значення:")
print(greater_than_mean)

# Обчислення добутку цих елементів
product_of_elements = np.prod(greater_than_mean)
print("Добуток елементів, які більше за середнє арифметичне значення:")
print(product_of_elements)
```

```python
# Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, 
# що містить найбільше за модулем ціле значення в кожному рядку матриці.
import numpy as np

# Створення матриці 3x3 з випадковими дійсними числами від -0.5 до 0.5
matrix = np.random.rand(3, 3) - 0.5
print("Матриця 3x3:")
print(matrix)

# Знаходження найбільшого за модулем цілого значення в кожному рядку
max_abs_int_vector = np.array([int(np.round(np.max(np.abs(row)))) for row in matrix])
print("Вектор-стовпчик з найбільшими за модулем цілими значеннями в кожному рядку:")
print(max_abs_int_vector)
```
